## <font color='blue'>Controlling your turtle</font>

You can control your turtle's <font color='red'>speed</font>, <font color='red'>position</font>, and whether it draws a <font color='red'>line</font> using a few extra commands.
Play around with different values below to control the turtle above.

```python

# t.speed(N) will control the speed of the turtle when it moves (N from 1 - 10)
t.speed(3)

# t.penup() will "raise the pen" meaning the turtle won't draw lines anymore
t.penup()


# And this will "lower the pen" so the turtle draws when it moves
t.pendown()

# This changes the color of the turtle's path, if the pen is down. default color is black
t.pencolor("red")  # try "blue", "yellow", "brown", "black", "purple", "green"


# This will rotate the turtle counter-clockwise
t.left(180)


# And this rotates the turtle clockwise
t.right(90)

# Move the Turtle forward by num units.

t.forward(50)

# Try: move the turtle forward by nagative num units.

t.forward(-100)

# Move the Turtle backward by num units.

t.backward(100)

# Draw a circle, or part of a circle. If radius is positive, it will turn to its left; a negative radius will make it turn to its right.

t.circle(100)



```

```python
# Desing your own circle function

def draw_circle(turtle,color,size, x,y):
    turtle.speed(3)
    turtle.penup()
    turtle.setposition(x, y)
    turtle.pendown()
    turtle.pencolor(color)
    turtle.circle(size)
    
```

```python
t3 = Turtle()

t3.speed(3)

draw_circle(t3,"yellow",30,100,100)

draw_circle(t3,"blue",30,200,200)

draw_circle(t3,"red",30,300,300)
```

```python
# Define your own circle function. Play the funciton. 

# Operation Add "+". two numbers.Join two strings together.

def draw_circle_Yang(turtle,color,size, x,y):
    turtle.speed(3)
    turtle.penup()
    turtle.setposition(x, y)
    turtle.pendown()
    turtle.pencolor(color)
    turtle.circle(size)
    turtle.circle(size+5)
 
```

```python

t_Yang = Turtle()

t_Yang.speed(3)

draw_circle_Yang(t_Yang,"yellow",30,150,150)

draw_circle_Yang(t_Yang,"blue",-30,150,250)

draw_circle_Yang(t_Yang,"red",30,250,150)

draw_circle_Yang(t_Yang,"green",-30,250,250)
```

```python
# For loop in Python

t4 = Turtle()
t4.speed(5)
colours=["red","blue","yellow","brown","black","purple","green"]

t4.penup(); t4.left(90); t4.forward(200);t4.right(90);t4.pendown()
for i in range (0,12):
    t4.pencolor(colours[i%7])
    t4.right(30)
    t4.forward(50)


t4.right(180)
t4.penup()
t4.home()
```
